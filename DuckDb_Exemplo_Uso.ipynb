```python
import duckdb as dkdb
import pandas as pd
```

```python
# Create a sample pandas DataFrame
data = {
    'id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [24, 30, 35, 29, 22],
    'city': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Houston']
}
df_sample = pd.DataFrame(data)
```

```python
query = (
"""
SELECT name, age
FROM df_sample
WHERE age > 25
ORDER BY age DESC
"""
)

df = dkdb.sql(query).df()
```
